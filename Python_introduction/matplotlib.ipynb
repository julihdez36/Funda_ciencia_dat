```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import random
x = list(range(1,8))
y = [random.randint(46,52) for i in range(7)]
y
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='green',linewidth = 3, linestyle = 'dotted')
```

```python
# Format strings in plot function
plt.plot(x,y,'g*--') #g of green, follow by shape
plt.plot(x,y,'rD')
plt.show()
```

```python
# If I want to specify arguments

plt.plot(x,y, color = 'red', marker = 'D', linestyle = '',markersize = 20,alpha = .3)
plt.show()
```

```python
# Axes labels, legend and grid

days = x
t_1 = y
t_2 = [random.randint(40,46) for i in range(7)]
t_3 = [random.randint(40,52) for i in range(7)]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, t_1, label = 'Temperature 1')
plt.plot(days, t_2, label = 'Temperature 2')
plt.plot(days, t_3, label = 'Temperature 3')

plt.legend(loc = 'lower right', shadow = True) # loc = 'best','upper left','center left'
plt.grid()
plt.show()
```

```python
# Bar chart

company = ['Google','Amazon','Microsoft','Facebook']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company) #position,replace
plt.xlabel('revenue(bln)')
plt.title('US tech stock')
plt.bar(xpos-0.2,revenue, width=0.4, label = 'Revenue')
plt.bar(xpos+0.2,profit, width=0.4, label = 'Profit')
plt.legend()

#plt.barh() barras horizontales
```

```python
# Histograms
variable_1 = [random.randint(100,130) for i in range(100)]
variable_2 = [random.randint(90,120) for i in range(100)]
```

```python
plt.title('Histogram example')
plt.ylabel('Total of values')
plt.xlabel('Indexing')
plt.hist([variable_1,variable_2], bins=3, rwidth= 0.95, color = ['g','orange'],
         label=['Variable 1','Variable 2'],orientation='horizontal')
plt.legend()
plt.show()
```

```python
# Pie chart
vals = [random.randint(200,1000) for i in range(5)]
labels = [f'label {i}' for i in range(5)]
```

```python
plt.axis('equal')
plt.pie(vals,labels = labels,radius=0.7, autopct='%.2f%%',explode=[0,0.1,0.2,0,0],
        startangle=360)
plt.show()
```

```python
np.random.seed(19680801)  # seed the random number generator.
data = {'a': np.arange(50), #Rengo de numpy
        'c': np.random.randint(0, 50, 50), #aleatorio int de numpy
        'd': np.random.randn(50)} #distribución normal
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100
```

```python
fig, ax = plt.subplots(figsize = (5,2.7), layout='constrained') #unidades en pulgadas, layoud, minimiza el espacio
ax.scatter('a','b',c = 'c',s = 'd', data = data) #color and size
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
plt.show()
```
