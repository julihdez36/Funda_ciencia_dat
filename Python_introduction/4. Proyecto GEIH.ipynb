### Funcion lambda
Una función lambda en Python es una función anónima y sin nombre que se puede definir en línea. A diferencia de las funciones normales definidas con la palabra clave def, una función lambda se crea utilizando la palabra clave lambda.

lambda arguments: expression

arguments son los argumentos de la función separados por comas.
expression es una expresión que define la operación a realizar en función de los argumentos.

ejemplo:
sumar = lambda x, y: x + y

resultado = sumar(3, 5)
print(resultado)  # Output: 8

```python
# OLS

import statsmodels.api as sm
```

```python
# Datos de ejemplo
x = df[['ESC']]        # Variable independiente como DataFrame
y = df['LnINGLABO']    # Variable dependiente como Serie

# Agregar una columna de unos para el intercepto
x = sm.add_constant(x)

# Crear el modelo de regresión lineal
model = sm.OLS(y, x)

# Ajustar el modelo a los datos
results = model.fit()

# Imprimir el resumen de la regresión
print(results.summary())
```
